```python
import pandas as pd
import yfinance as yf
import re
```

```python
# Load in data and create "tickers" list
sp500_df = pd.read_csv('sp500_cik.csv')
tickers = sp500_df['Symbol'].dropna().unique().tolist()

# Compile all data in "sp500_data" dataframe
sp500_data = pd.DataFrame()
start_date = "2001-01-01"
end_date = "2025-01-01"

for ticker in tickers:
    data = yf.download(ticker, start=start_date, end=end_date, group_by='ticker')

    if isinstance(data.columns, pd.MultiIndex):
        df = data[ticker].copy()
    else:
        df = data.copy()

    df = df.reset_index()
    df['Ticker'] = ticker
    df = df[['Date', 'Ticker', 'Open', 'High', 'Low', 'Close', 'Volume']]

    sp500_data = pd.concat([sp500_data, df], ignore_index=True)

sp500_data.columns.name = None

# Export csv
sp500_data.to_csv("sp500_data.csv", index=False)
```
